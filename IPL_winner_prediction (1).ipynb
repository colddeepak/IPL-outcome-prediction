**Creating new columns Current_Score,runs_left, balls_left,wicket_left current_run_rate, required_run_rate,result**

```python
balls['current_score'] = balls.groupby('ID')['total_run'].cumsum()
```

```python
balls
```

```python
balls['runs_left']=np.where(balls['target'] - balls['current_score']>=0, balls['target'] - balls['current_score'],0)
```

```python
balls
```

```python
balls['balls_left'] = np.where(120 - balls['overs']*6 - balls['ballnumber'] >= 0, 120-balls['overs']*6 - balls['ballnumber'], 0)
```

```python
balls['wickets_left'] = 10 - balls.groupby('ID')['isWicketDelivery'].cumsum()
```

```python
balls.columns
```

```python
balls['current_run_rate'] = (balls['current_score']*6) / (120-balls['balls_left'])
```

```python
balls['required_run_rate'] = np.where(balls['balls_left']>0, balls['runs_left']*6/balls['balls_left'],0)
```

```python
balls.columns
```

```python
def result(row):
  return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
balls['result'] = balls.apply(result, axis=1)
```

```python
balls.head()
```

```python
index1 = balls[balls['Team2']==balls['BattingTeam']]['Team1'].index
index2 = balls[balls['Team1']==balls['BattingTeam']]['Team2'].index
```

```python
balls.loc[index1, 'BowlingTeam'] = balls.loc[index1, 'Team1']
balls.loc[index2, 'BowlingTeam'] = balls.loc[index2, 'Team2']
```

**Transformation of columns containing categorical value into numerical value**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','TossWinner','TossDecision','City'])
],
remainder = 'passthrough')
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('result', axis=1)
Y = df['result']
X.shape, Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression())
])
```

```python
pipe.fit(X_train, Y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, Y_test)
```

```python
pipe.predict_proba(X_test)
```

```python
DT_pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', DecisionTreeClassifier(criterion = 'gini', max_depth = 4, min_samples_split = 20, random_state = 42))
])
```

```python
DT_pipe.fit(X_train, Y_train)
```

```python
DT_Ypred = DT_pipe.predict(X_test)
```

```python
accuracy_score(DT_Ypred, Y_test)
```
